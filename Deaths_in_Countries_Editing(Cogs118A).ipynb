```python
labels = []

for i in list(df):
    if i.find('-') == -1:
        labels.append(i)
    else:
        label = i.split(' - ')[1].strip()
        labels.append(label)
print(labels)
```

```python
def continent_code(country_code):
    try:
        alpha2 = pc.country_alpha3_to_country_alpha2(country_code)
        continent_code = pc.country_alpha2_to_continent_code(alpha2)
        return continent_code
    except:
        return np.nan
```

```python
has_country_code['Continent_Code'] = has_country_code['Code'].apply(continent_code)
```

```python
to_save = to_save[['Entity', 'Code', 'Continent_Code'] + [i for i in to_save if i not in ['Entity', 'Code', 'Continent_Code']]]
```

```python
subset_columns = ['Number of executions (Amnesty International)', 'Terrorism (deaths)']
```

```python
to_save = to_save.dropna(subset = [i for i in to_save if i not in subset_columns])
```
